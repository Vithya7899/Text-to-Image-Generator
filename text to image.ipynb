```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")
```

```python
style_map = {
    "Realistic": "ultra-realistic, 8k, sharp focus, high detail",
    "Anime": "anime style, highly detailed, vibrant colors",
    "Cartoon": "cartoon style, simplified, colorful, clean lines",
    "Oil Painting": "oil painting style, brushstrokes, textured canvas",
    "Cyberpunk": "cyberpunk, futuristic, neon lights, high detail"
}
```

```python
import gradio as gr

def generate_image(prompt, style):
    styled_prompt = f"{prompt}, {style_map.get(style, '')}"
    image = pipe(styled_prompt).images[0]
    return image

styles = list(style_map.keys())

with gr.Blocks() as demo:
    gr.Markdown("# ðŸŽ¨ Text to Image Generator")

    with gr.Row():
        prompt_input = gr.Textbox(label="Enter your prompt", placeholder="e.g., A sunset over a mountain range")
        style_dropdown = gr.Dropdown(choices=styles, label="Choose Style", value="Realistic")

    generate_button = gr.Button("Generate Image")
    output_image = gr.Image(label="Generated Image", type="pil")

    generate_button.click(fn=generate_image, inputs=[prompt_input, style_dropdown], outputs=output_image)

demo.launch(share=True)
```
